```python
Q1. Create a function which will take a list as an argument and return the product of all the numbers
after creating a flat list.
Use the below-given list as an argument for your function.
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']


ANS== I create a recursive function that flattens the given list and calculates the product of all the numbers found in the flattened list.
CODE:
    
def flatten_and_calculate_product(lst):
    flattened_list = []
    for item in lst:
        if isinstance(item, (list, tuple, set)):
            flattened_list.extend(flatten_and_calculate_product(item))
        elif isinstance(item, dict):
            flattened_list.extend(flatten_and_calculate_product(item.keys()))
            flattened_list.extend(flatten_and_calculate_product(item.values()))
        elif isinstance(item, (int, float)):
            flattened_list.append(item)
    
    product = 1
    for num in flattened_list:
        product *= num
    
    return product

list1 = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34), {1, 2, 3, 3, 2, 1}, {1: 34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']

result = flatten_and_calculate_product(list1)
print("Product:", result)

output: 
    Product: 42199216640

    In this code, I define the function flatten_and_calculate_product that takes a list as an argument. 
    Inside the function, i initialize an empty list flattened_list to store all the numeric elements found in the list.
    
```
